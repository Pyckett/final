2. Implement an algorithm to determine if a string has all unique
characters. Do not use a set.

```python
def isUnique(string):
    seen = []
    # simple create a seen array, if in seen then the whole algo "fails"
    for c in string:
        if c in seen:
            return False
        
        if c not in seen:
            seen.append(c)
    
    return True
```

8. How would you design a stack which, in addition to push and pop,
has a function called min that returns the minimum element? You
may use an additional stack.

```python
class MinStack():
    def __init__(self):
        # This stack holds all values in the order they were pushed
        self.main = []

        # This stack only tracks minimum values seen so far
        # The top of this stack is always the current minimum
        self.min_stack = []

    def push(self, value):
        # Always push the balue onto the main stack
        self.main.append(value)
        # If minstack empty or the last value is less than the last element of minstack
        # we push it on onto min_stack so it becomes the new minimum
        if not self.min_stack or value <= self.min_stack[-1]:
            self.min_stack.append(value)

    def pop(self, value):
        # Remove and save top value from main stack
        popped = self.main.pop()

        # If the value we just removed is the current minimum,
        # we also pop it from min_stack to expose the previous minimum
        if value == self.min_stack[-1]:
            self.min_stack.pop()
        
        # return our stored removed value
        return popped
    
    def min(self):
        # since top of min_stack is always the minimum return the last element in the list
        # O(1) obviously
        return self.min_stack[-1]
```

23. Write an algorithm to find the k-th largest element in an unsorted
array without using heapq.

```python
def kthlargest(array, k):
    # how to remember insertion sort by memory
    # grow a sorted list one element at a time by inserting the next item into its correct position
    
    # this is the growing sorted list
    new_array = []

    # for every number
    for x in array:
        # obviously new_array needs to be populated
        if not new_array:
            new_array.append(x)
            continue
        
        # inserted setup flag
        inserted = False

        # I need to know if it's found its position yet
        # so we scan left to right until I find something bigger
        # insert before it by saving the i.
        for i in range(len(new_array)):
            if x < new_array[i]:
                new_array.insert(i,x)
                inserted = True
                break
        
        # if nothing found then add it to the end
        if not inserted:
            new_array.append(x)
    
    # then return the kth element which is -k
    return new_array[-k]

def kthlargest2(array, k):
    array.sort()
    return array[-k]
```

30. Write an algorithm to check if two binary trees are identical.

```python
# I can do this one easily
def same_tree(p, q):
    if not p and not q: # this is the base case
        return True
    
    if not p or not q: # since the first one didnt trigger, it's obviously not the base case
        return False

    if p.val != q.val: # needs to be value, checking p and q directly is wrong
        # it's wrong because otherwise we are checking two objects in memory
        # which can have the same value and not equal each other
        return False
    
    #compares left subtree
    left_same = same_tree(p.left, q.left)
    
    #compares right subtree
    right_same = same_tree(p.right, q.right)

    # for a diagram it'll go all the way down the left, then the parts of that subtree.
    # and only when its fully done will it check the right subtree of the root.
    return left_same and right_same
```
